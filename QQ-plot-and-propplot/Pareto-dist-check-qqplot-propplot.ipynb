# Pareto distribution Check

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
# generate non normal distribution data (exponential data distribution)
original_data = np.random.exponential(size=1000)
original_data
```

```python
# transform training data and save lambda value
# Here were are Passing Pareto Distribution to Boxcox transform for Transformation
fitted_data,fitted_lambda = stats.boxcox(original_data)
```

```python
fitted_data
```

```python
fitted_lambda
```

```python
fig,ax = plt.subplots(1,2)
```

```python
# plotting the original data
sns.distplot(original_data,hist=False,kde=True,label="Non-normal",color="green",ax=ax[0])
#plt.legend(loc="upper right")

sns.distplot(fitted_data,hist=False,kde=True,label="Normal",color="green",ax=ax[1])
#plt.legend(loc="upper right")

```

```python
fig
```

### Here we can see that the Pareto distribution is transformed into Standard Normal Distribution


```python
sm.qqplot(fitted_data,line='45')
```

## This is not a Pareto Distribution
## Here we can see even the transformed data distribution is not Guassian distributed. If the Distribution was Guassian then we could have said it is a Pareto Distribution
