```python
import numpy as np
import statsmodels.api as sm
import pylab as py
import seaborn as sns
```

# QQ Plot to check if the distribution is a Guassian Distribution(Standard Normal Distribution) or not using QQ-plot

```python
# Create Guassian distribution data
data_points = np.random.normal(0,1,100) # Create a normal distribution
print(data_points)
sns.distplot(data_points,kde=True)
sm.qqplot(data_points,line='45') # check if its a normal dist using qq plot
py.show()
```

## Here since majority of data_points values lie on the staright line(ideal line for standard normal distribution) we say this is a Guassian Distribution

## Another way to check using probplot

```python
import scipy.stats as stats
import pylab
import matplotlib.pyplot as plt
```

```python
samples = np.random.normal(0,1,100) # create a Normal distribution
stats.probplot(samples,dist='norm',sparams=(15,10),plot=pylab) # check if its a normal distribution using probplot
plt.show()
```

# Log Normal Distribution check

```python
# Create log normal distribution data
values = np.random.lognormal(0,1,100)
print(values)
sns.distplot(values,kde=True)
sm.qqplot(values,line='45')
py.show()
```

## Here since majority of values does not lie on the staright line(ideal line for standard normal distribution) we say this is not a Guassian Distribution , also in Log normal distribution the distribution is skewed to once side unlike guassian which is almost symmetrical on both sides

# To check if this Distribution is Log Normal Distribuution

```python
sm.qqplot(np.log(values),line='45')
py.show()
```

## Here since majority of values lie on the staright line(ideal line for log normal distribution) we say this is a Log Normal Distribution
