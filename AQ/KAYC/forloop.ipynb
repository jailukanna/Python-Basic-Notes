```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
yelp_url = "https://www.yelp.com/search?cflt=assistedliving&find_loc=Oakland+County%2C+MI"
#yelp_url = "https://www.yelp.com/search?cflt=assistedliving&find_loc=Oakland%20County%2C%20MI&start=10"
#yelp_url = "https://www.yelp.com/search?cflt=assistedliving&find_loc=Oakland%20County%2C%20MI&start=20"

uClient = uReq(yelp_url)
yelpPage = uClient.read()
yelpPage
#uClient.close()
```

```python
yelp_html = bs(yelpPage, "html.parser")
yelp_html
```

```python
bigboxes = yelp_html.findAll("div", {"class": "businessName__09f24__3Ml2X display--inline-block__09f24__3SvIn border-color--default__09f24__3Epto"})

print(bigboxes)
```

```python
# Create a csv file
filename = "assisted_living_links" + ".csv"
fw = open(filename, "w")
#headers = "Facility Link, Facility Name, Website, Contact Number, Address \n"
#fw.write(headers)
reviews = []
```

```python
#Create a Column name
import csv
header = [['Facility Link', 'Facility Name', 'Website', 'Contact Number', 'Address']]
file = open('assisted_living_links.csv','w+',newline='')
with file:
    write = csv.writer(file)
    write.writerows(header)
```

```python
for i in range(len(bigboxes)):
    #print(i,bigboxes[i])
    facilityLink = "https://www.yelp.com" + bigboxes[i].div.a['href']
    print("\n")
    print(i+1,'.',facilityLink)
    facilityName = bigboxes[i].div.find_all('a', {'class': 'css-166la90'})[0].text
    print("Facility name:",facilityName)
    facilityResponse = requests.get(facilityLink)
    facilityResponse
    facilityResponse.encoding='utf-8'
    facility_html = bs(facilityResponse.text, "html.parser")
    infoBoxes = facility_html.findAll('div',{'class':"arrange__373c0__2iVWK gutter-2__373c0__1a5Xm vertical-align-middle__373c0__2sr2a border-color--default__373c0__r305k"})
    try:
        place_website = infoBoxes[0].find_all('p', {'class': 'css-1h1j0y3'})[0].text
        print("Website:",place_website)
    except:
        print('Nan')
        place_website='Nan'
    try:
        phone_number = infoBoxes[1].find_all('p', {'class': 'css-1h1j0y3'})[0].text
        print("Contact Number:",phone_number)
    except:
        print('Nan')
        phone_number='Nan'
    try:
        address = infoBoxes[2].find_all('p', {'class': 'css-chtywg'})[0].text
        print("Address:",address)
    except:
        print('Nan')
        address='Nan'
    
    all_info = [facilityLink,facilityName,place_website,phone_number,address]
    reviews.append(all_info)
```

```python
reviews
```

```python
# Append all the information to the already created csv file
import csv
file = open('assisted_living_links.csv','a+',newline='')
with file:
    write = csv.writer(file)
    write.writerows(reviews)
```
