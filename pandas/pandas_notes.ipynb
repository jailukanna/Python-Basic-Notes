```python
res = requests.get('https://api.github.com/repos/pandas-dev/pandas/issues')
```

```python
res
```

```python
js = res.json() # This has very complex json format
js #we can see that it is inside a list and in dict format
```

```python
js[0]['title'] #find the value title key in first item of the list
```

```python
for i in range(len(js)):
    print(js[i]['title'])
#find the value title key all of the list
```

```python
selected = pd.DataFrame(js,columns=['url','id','title']) #selecte which all columns we need
selected
```

```python
all_columns = pd.DataFrame(js)
all_columns
```

```python
all_columns.to_csv("json_dump.csv",index=False)
```
