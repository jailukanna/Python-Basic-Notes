# Visualization

## Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
np.random.seed(20)
df = pd.DataFrame(np.random.rand(1000,1), index=pd.date_range('2021-07-18',periods = 1000),columns=['data'])
df
```

# 1. Plot()

```python
df.plot()
```

# 2. Cumsum (Cumulative sum)
#  Cumprod (Cumulative product)

```python
df.data.cumsum() #cumulative sum
```

```python
df.data.cumsum().plot()
```

```python
df.data.cumprod().plot()
```

```python
df.data.cummax().plot()
```

```python
df['Cumsum'] = df.data.cumsum()
```

```python
df['Cubed'] = df['data'].apply(lambda x: x**3)
```

```python
df
```

```python
df.plot()
```
